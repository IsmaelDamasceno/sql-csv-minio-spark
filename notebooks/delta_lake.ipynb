```python
from projeto_ed_satc.spark_session import get_spark

spark = get_spark()
```

```python
df = spark.createDataFrame([(1, "Alice"), (2, "Bob")], ["id", "name"])
df.write.format("delta").mode("overwrite").save("warehouse/delta/people")

spark.read.format("delta").load("warehouse/delta/people").show()
```

```python
from delta.tables import DeltaTable

dt = DeltaTable.forPath(spark, "warehouse/delta/people")
dt.update("id = 1", {"name": "'Alicia'"})

spark.read.format("delta").load("warehouse/delta/people").show()
```

```python
spark.read.format("delta").option("versionAsOf", 0).load("warehouse/delta/people").show()
```
